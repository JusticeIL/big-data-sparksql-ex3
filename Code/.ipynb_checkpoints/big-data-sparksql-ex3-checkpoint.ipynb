```python
from pyspark.sql import SparkSession
```

```python
# Create a SparkSession
spark = SparkSession.builder \
    .appName("big-data-sparksql-ex3") \
    .getOrCreate()
```

```python
BASE_ABSOLUTE_PATH = "/Users/galrubinstein/Documents/תואר ראשון מדמ״ח/ביג דאטה/SparkSQL Assignment/pyspark/simpsons/"
```

```python
df_lines = spark.read.format("com.databricks.spark.csv").options(header='true') \
.option("delimiter", ",").option("quote", '"') \
.option("escape", '"').option("multiLine", True) \
.load(BASE_ABSOLUTE_PATH + "simpsons_script_lines.csv")
df_lines.createOrReplaceTempView("lines_table")
```

```python
## print lines file schema
df_lines.printSchema()
```

```python
## print all fields for first 5 rows
spark.sql("SELECT * from lines_table").show(5)
```

```python
## print the shortest 10 script lines:
spark.sql("SELECT int(word_count) AS wc, raw_text FROM lines_table WHERE TRY_CAST(word_count AS INT) > 0 ORDER BY wc ").show(10,False)
```

```python
## count the scenes that took more than 10 minutes
spark.sql("SELECT count(*) FROM lines_table WHERE TRY_CAST(timestamp_in_ms AS INT) > (10 * 60 * 1000)").show()
```

## print the first 50 locations that has the word "Springfield" in them , ignoring letters case.


```python
## ## write your code here ##
df_locations = spark.read.format("com.databricks.spark.csv").options(header='true') \
.option("delimiter", ",").option("quote", '"') \
.option("escape", '"').option("multiLine", True) \
.load(BASE_ABSOLUTE_PATH + "simpsons_locations.csv")
df_locations.createOrReplaceTempView("locations_table")

spark.sql("SELECT name FROM locations_table WHERE normalized_name like '%springfield%'").show(50, False)
```

## print 20 quotes that are located in any place that has Jerusalem in its name.

```python
## Note that Jerusalem may appear in any case and in any part of the location name.
## Use JOIN for this query on another table
## ## write your code here ##
spark.sql("SELECT spoken_words AS quotes \
FROM lines_table \
INNER JOIN locations_table \
ON lines_table.raw_location_text = locations_table.name \
WHERE lines_table.speaking_line = 'true' \
AND locations_table.normalized_name like '%jerusalem%'") \
.show(20, False)
```

## print first 20 most used locations with the count of lines spoken in them.

```python
## use GROUP BY for that
## ## write your code here ##
spark.sql("SELECT locations_table.name AS locations, COUNT(*) AS number_of_lines \
FROM lines_table \
INNER JOIN locations_table \
ON lines_table.raw_location_text = locations_table.name \
WHERE lines_table.speaking_line = 'true' \
GROUP BY locations_table.name \
ORDER BY number_of_lines DESC") \
.show(20, False)
```

## find the seasons in which the average imdb rating was the highest.

```python
## ## write your code here ##
df_episodes = spark.read.format("com.databricks.spark.csv").options(header='true') \
.option("delimiter", ",").option("quote", '"') \
.option("escape", '"').option("multiLine", True) \
.load(BASE_ABSOLUTE_PATH + "simpsons_episodes.csv")
df_episodes.createOrReplaceTempView("episodes_table")
```

```python
spark.sql("SELECT season, AVG(imdb_rating) AS avg_rating \
FROM episodes_table \
GROUP BY season \
HAVING avg_rating = ( \
        SELECT MAX(avg_season_rating) \
        FROM ( \
            SELECT AVG(imdb_rating) AS avg_season_rating \
            FROM episodes_table \
            GROUP BY season \
        ))").show(df_episodes.count(), False)
```

## Print the seasons number, the number of episodes in each one and the average rating
## in a descending order from highest average rating to lowest.

```python
spark.sql("SELECT season, COUNT(*) as episodes_count, AVG(imdb_rating) AS avg_rating \
FROM episodes_table \
GROUP BY season \
ORDER BY avg_rating DESC") \
.show(df_episodes.count(), False)
```
